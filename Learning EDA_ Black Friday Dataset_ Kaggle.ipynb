# Removing a Non-Useful Column

```python
df_concat.drop('User_ID', axis=1, inplace=True)
#axis=1 for vertical removal, axis=0 for horizontal removal
#inplace keeps the rest of the dataset intact and the next column is assigned the index values.
```

```python
df_concat.head()
```

Observe the categorical variables and make the dataset ready
Clearly gender is a categorical variable or we can use unique function


```python
df_concat['Gender'].unique()
```

We need to set integer values to these categorical variable-- using the map function

```python
df_concat['Gender']= df_concat['Gender'].map({'F':0, 'M':1})
# Observation: When performing mapping of such categorical variables the order of variables is important.
#     If I write map({'M':1, 'F':0}) it maps NaN values in Gender column
```

```python
df_concat.head()
```

```python
df_concat['Age'].unique()
```

```python
df_concat['Age']=df_concat['Age'].map({'0-17':1, '18-25':2 ,'26-35':3,'36-45':4, '46-50':5, '51-55':6,'55+':7 })
#This is technically called as - Target ordinal encoding
# Incase the categories are more and they cannot be written and mapped -- use label encoding - (sklearn library - preprocessing-- Google and refer the code in geekforgeeks website)
```
