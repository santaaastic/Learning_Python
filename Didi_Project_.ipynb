```python
#From Github
# Credits: https://medium.com/towards-entrepreneurship/importing-a-csv-file-from-github-in-a-jupyter-notebook-e2c28e7e74a5
import pandas as pd
import requests
import io

url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv" # Make sure the url is the raw version of the file on GitHub

download = requests.get(url).content

# Reading the downloaded content and turning it into a pandas dataframe

df = pd.read_csv(io.StringIO(download.decode('utf-8')))

# Printing out the first 5 rows of the dataframe

print (df.head())
```
